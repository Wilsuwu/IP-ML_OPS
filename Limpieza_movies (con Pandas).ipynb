```python
#
#   Importacion de librerias
#
#!pip install pandas
#!pip install numpy
import pandas as pd
import numpy as np
```

```python
#Leyendo archivo.csv

movies_df = pd.read_csv('movies_dataset.csv', low_memory=False)
```

```python
movies_df.head()
```

```python
# Dropeo columnas solicitadas
# los pasos no se encuentran en el orden solicitado, sino que los fui haciendo de acuerdo a como más comodo me quedará trabajar luego los siguientes pasos

#creamos una lista con las columnas a dropear
droplist = ['video','imdb_id','adult','original_title','vote_count','poster_path','homepage']

#dropeamos
movies_df = movies_df.drop(columns=droplist)

```

```python
# chequeamos la cantidad de valores nulos

print(movies_df.isna().sum())
```

```python
# cambiamos nulos por el número 0

movies_df['revenue'].fillna(0, inplace=True)
```

```python
# volvemos a chequear luego de hacer el cambio

print(movies_df.isna().sum())
```

```python
movies_df.info()
```

```python
# Encontramos el error que no permitia la transformacion del tipo de dato en la columna 'budget' y dropeamos dicha fila

#movies_df[movies_df['budget'] == '/ff9qCepilowshEtG2GYWwzt2bs4.jpg']
#movies_df[movies_df['budget'] == '/zV8bHuSL6WXoD6FWogP9j4x80bL.jpg']

#movies_df = movies_df.drop(movies_df[movies_df['budget'] == '/ff9qCepilowshEtG2GYWwzt2bs4.jpg'].index)


# Al notar que no es solo un dato el que impide la transformacion procedo a buscar con ayuda de chatgpt una forma eficiente de 
# filtrar las mismas y convertir el tipo de dato para realizar la transformacion necesaria

# Creamos un filtro
filtro = movies_df['budget'].str.contains('.jpg')

#
# aplicamos el filtro generado con el operador de negacion '~' para invertir el mismo y obtener las columnas que no contengan '.jpg'

movies_df = movies_df[~filtro]

# Finalmente convertimos el tipo de dato
movies_df['budget'] = movies_df['budget'].astype('float64')


```

```python
# Creamos columna return

movies_df['return'] = movies_df['revenue'] / movies_df['budget']
```

```python
# Reemplazamos los NaNs e infinitos en la columna return

movies_df['return'] = movies_df['return'].replace(np.inf, 0)
movies_df['return'] = movies_df['return'].fillna(0)
```

```python
# Transformamos el tipo de dato del campo 'release_date'
movies_df['release_date'] = pd.to_datetime(movies_df['release_date'], format= '%Y-%m-%d')
```

```python
# Tratamos nulos
movies_df['release_date'].isna()
```

```python
# Eliminamos nulos de la columna 'release_date'
movies_df['release_date'] = movies_df['release_date'].dropna()
```

```python
movies_df.info()
```

```python
#CREAMOS LA COLUMNA RELEASE YEAR
movies_df['release_year'] = movies_df.release_date.dt.year
```

```python
movies_df['release_year'].isna()
```

```python
# Reemplazamos los NaNs e infinitos en release_year

movies_df['release_year'] = movies_df['release_year'].replace(np.inf, 0)
movies_df['release_year'] = movies_df['release_year'].fillna(0)
```

```python
# Transformamos el tipo de dato ya que fue tomado como float64 en un principio
movies_df['release_year'] = movies_df['release_year'].astype('int')
```

```python
# Instanciamos la funcion para desanidar
import ast

def fetch_name(obj): # Está funcion tiene el proposito de extraer una lista de nombres de una cadena de texto si esta contiene un diccionario
    if isinstance(obj, str) and '{' in obj:
        L=[]
        for i in ast.literal_eval(obj):
            L.append(i['name']);
        return L
```

```python
def fetch_name2(obj): #Está funcion extrae el valor asociado a la clave 'name'
    if isinstance(obj, str) and '{' in obj:
        # print(obj)
        dic = ast.literal_eval(obj)
        return dic['name']
```

```python
movies_df
```

```python
#Expresamos los datos de las columnas que tienen datos en forma de listas a forma de no listas
movies_df['genres'] = movies_df['genres'].apply(fetch_name)
movies_df['belongs_to_collection'] = movies_df['belongs_to_collection'].apply(fetch_name2)
movies_df['production_companies'] = movies_df['production_companies'].apply(fetch_name)
movies_df['production_countries'] = movies_df['production_countries'].apply(fetch_name)
movies_df['spoken_languages'] = movies_df['spoken_languages'].apply(fetch_name)
```

```python
# Notamos que aún permanecen los '[]', así que procedemos a transformar eso
movies_df.head(10)
```

```python
# Usamos este codigo que iterara cada columna y nos devolvera una cadena de strings separados por ',', en el que si no hay valor conservara el valor None.
# La misma hará que los datos sean más legibles y faciles de analizar


movies_df['genres'] = movies_df['genres'].apply(lambda x: ', '.join(map(str, x)) if x is not None else None)
movies_df['production_companies'] = movies_df['production_companies'].apply(lambda x: ', '.join(map(str, x)) if x is not None else None)
movies_df['production_countries'] = movies_df['production_countries'].apply(lambda x: ', '.join(map(str, x)) if x is not None else None)
movies_df['spoken_languages'] = movies_df['spoken_languages'].apply(lambda x: ', '.join(map(str, x)) if x is not None else None)
```

```python
# Comprobamos que ya no se encuentran los mismos
movies_df.head(10)
```

```python
#Transformamos el tipo de dato de la columna 'id' para poder iterarlo en una funcion

movies_df['id'] = movies_df['id'].astype(int)

print(movies_df.dtypes)
```

```python
# Contar la cantidad de valores NaN o NA en la columna "production_countries"
cantidad_nulos = movies_df["production_countries"].isna().sum()

print(f"Cantidad de valores nulos: {cantidad_nulos}")
```

```python
# Cambiamos las filas que contienen NaN en 'production_countries', ya que ocasiona problemas al momento de crear las funciones
#movies_df['production_countries'] = movies_df['production_countries'].fillna('Unknown')

movies_df['production_companies'] = movies_df['production_companies'].fillna('Unknown')
```
